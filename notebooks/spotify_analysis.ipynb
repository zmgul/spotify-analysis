# ðŸŽ§ Understanding Spotify Listeners: Music Preferences by Age and Gender

This project analyzes music genre preferences across different age groups and genders using a synthetic dataset.
The dataset includes age group, gender and preferred music genre of Spotify users.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("../data/spotify_listener_segments.csv")
df.head(10)
```

```python
# Music genre preferences by age group
plt.figure(figsize=(10, 6))
age_groups = ["13-17", "18-24", "25-34", "35-44", "45-54", "55+"]
sns.countplot(data=df, x="age_group", hue="preferred_genre", order=age_groups)

plt.title("Music Genre Preferences by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Music genre preferences by gender
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="gender", hue="preferred_genre", order=["Female", "Male"])

plt.title("Music Genre Preferences by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
